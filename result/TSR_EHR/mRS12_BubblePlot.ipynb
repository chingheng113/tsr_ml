```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = os.path.join("..", "..", "data","LINKED_DATA", "TSR_EHR", "TSR_12_CLEANED.csv")
tsr_12 = pd.read_csv(csv_path)
tsr_12.head()
```

```python
mrs_tx_1 = tsr_12.mrs_tx_1
mrs_tx_3 = tsr_12.mrs_tx_3
mrs_tx_6 = tsr_12.mrs_tx_6
mrs_tx_12 = tsr_12.mrs_tx_12
total = tsr_12.total
total_in = tsr_12.total_in
total_out = tsr_12.total_out
crp_nm = tsr_12.crp_nm
gcsv_nm = tsr_12.gcsv_nm
ldl_nm = tsr_12.ldl_nm
hospitalised_time = tsr_12.hospitalised_time
```
