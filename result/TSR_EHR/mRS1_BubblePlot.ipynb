```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = os.path.join("..", "..", "data","LINKED_DATA", "TSR_EHR", "TSR_1_CLEANED.csv")
tsr_1 = pd.read_csv(csv_path)
tsr_1.head()
```

```python
mrs_tx_1 = tsr_1.mrs_tx_1
total = tsr_1.total
total_in = tsr_1.total_in
total_out = tsr_1.total_out
crp_nm = tsr_1.crp_nm
gcsv_nm = tsr_1.gcsv_nm
ldl_nm = tsr_1.ldl_nm
hospitalised_time = tsr_1.hospitalised_time
```

# mRS VERSUS suspected features

## 6 Outcomes

```python
fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (15,5))
ax1.scatter(mrs_tx_1, hospitalised_time, s = 100,  alpha=0.1)
sns.kdeplot(hospitalised_time,hue = mrs_tx_1,vertical=False, fill=True, alpha = 0.3, palette="crest", linewidth=1)
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (15,5))
ax1.scatter(mrs_tx_1, crp_nm, s = 100,  alpha=0.1)
sns.kdeplot(crp_nm,hue = mrs_tx_1,vertical=False, fill=True, alpha = 0.3, palette="crest", linewidth=1)
plt.xlim(0, 30)
plt.show()
```
