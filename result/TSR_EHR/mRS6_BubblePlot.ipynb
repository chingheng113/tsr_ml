```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = os.path.join("..", "..", "data","LINKED_DATA", "TSR_EHR", "TSR_6_CLEANED.csv")
tsr_6 = pd.read_csv(csv_path)
tsr_6.head()
```

```python
mrs_tx_1 = tsr_6.mrs_tx_1
mrs_tx_3 = tsr_6.mrs_tx_3
mrs_tx_6 = tsr_6.mrs_tx_6
total = tsr_6.total
total_in = tsr_6.total_in
total_out = tsr_6.total_out
crp_nm = tsr_6.crp_nm
gcsv_nm = tsr_6.gcsv_nm
ldl_nm = tsr_6.ldl_nm
hospitalised_time = tsr_6.hospitalised_time
```

```python
fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (15,5))
ax1.scatter(mrs_tx_6, hospitalised_time, s = 100,  alpha=0.1)
sns.kdeplot(hospitalised_time,hue = mrs_tx_6,vertical=False, fill=True, alpha = 0.3, palette="crest", linewidth=1)
plt.show()
```
