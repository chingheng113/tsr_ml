```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = os.path.join("..", "..", "data","LINKED_DATA", "TSR_EHR", "TSR_3_CLEANED.csv")
tsr_3 = pd.read_csv(csv_path)
tsr_3.head()
```

```python
mrs_tx_1 = tsr_3.mrs_tx_1
mrs_tx_3 = tsr_3.mrs_tx_3
total = tsr_3.total
total_in = tsr_3.total_in
total_out = tsr_3.total_out
crp_nm = tsr_3.crp_nm
gcsv_nm = tsr_3.gcsv_nm
ldl_nm = tsr_3.ldl_nm
hospitalised_time = tsr_3.hospitalised_time
```

```python
fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (15,5))
ax1.scatter(mrs_tx_3, hospitalised_time, s = 100,  alpha=0.1)
sns.kdeplot(hospitalised_time,hue = mrs_tx_3,vertical=False, fill=True, alpha = 0.3, palette="crest", linewidth=1)
plt.show()
```
