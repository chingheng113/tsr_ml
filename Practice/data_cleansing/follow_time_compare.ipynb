```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from matplotlib.font_manager import FontProperties
import numpy as np
import scipy.stats
```

```python
font_path = os.path.join("..", "kaiu.ttf")
myfont = FontProperties(fname= font_path, size=14)
sns.set(font = myfont.get_name())

plt.rcParams['figure.figsize'] = (10,5)
```

```python
csv_path = os.path.join("..", "data","TSR_CMUH", "follow_all_time.csv")
followalltime = pd.read_csv(csv_path)

followalltime.NIHSS_IN[(followalltime.NIHSS_IN < 0)|(followalltime.NIHSS_IN > 42)] = np.nan
followalltime.NIHSS_OUT[(followalltime.NIHSS_OUT < 0)|(followalltime.NIHSS_OUT > 42)] = np.nan
followalltime.BARTHEL[(followalltime.BARTHEL < 0)|(followalltime.BARTHEL > 100)] = np.nan

followalltime.head()
```
